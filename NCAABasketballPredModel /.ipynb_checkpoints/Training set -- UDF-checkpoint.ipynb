```python
def stats_per_season(TeamID, season): 
    
    #wins,losses, total games of each team 
    year_data = Stats[Stats['Season'] == season]
    
    GamesWon = year_data[year_data['WTeamID'] == TeamID]
    Gameslost = year_data[year_data['LTeamID'] == TeamID]
    
    NumWins = len(GamesWon.index)
    NumLoss = len(Gameslost.index)
    TotalGames = NumWins + NumLoss #missing values? 
    
    #seed of the team 
    #tourneyYear = Seeds[Seeds['Season'] == season]
    #seed = tourneyYear[tourneyYear['TeamID'] == TeamID]
    #seed = seed.values[0][1]
    #tourneySeed = int(seed[1:3])    

    # points scored and allowed 
    TotalScored = GamesWon['WScore'].sum()
    TotalScored += Gameslost['LScore'].sum()
    
    TotalAllowed = Gameslost['WScore'].sum()
    TotalAllowed += GamesWon['LScore'].sum()

    #gives us the ratings of the team per season 
    year_data = year_data.loc[(year_data.WTeamID == TeamID) | (year_data.LTeamID == TeamID), :]
    #since WteamID and LteamID equlas to TeamID, W + stats and L + stats will be the same
    
    eFG = ((year_data['WFGM']+ year_data['WFGM3']*.5)/ year_data['WFGA']).mean() 

    Poss = (year_data['WFGA'] - year_data['WOR']+ year_data['WTO'] +.475*year_data['WFTA'])

    TO_Percent = (year_data['WTO']/ Poss).mean() #turnover
    
    FTrate = (year_data['WFTA'] / year_data['WFGA']).mean()

    ORp = (year_data['WOR'] / (year_data['WOR'] + year_data['WDR'])).mean()
    
    WORtg = (GamesWon['WScore']/ Poss).mean() #points scored/ poss 
    LORtg = (Gameslost['LScore']/ Poss).mean()
    
    WDRtg = (GamesWon['LScore'] / Poss).mean()
    LDRtg = (Gameslost['WScore']/ Poss).mean()
    
    if NumWins == 0: 
        AdjORtg =(WORtg*0 + LORtg*NumLoss) / TotalGames 
        AdjDRtg = (WDRtg*0 + LDRtg*NumLoss) / TotalGames
    if NumLoss == 0: 
        AdjORtg =(WORtg*NumWins + LORtg*0) / TotalGames 
        AdjDRtg = (WDRtg*NumWins + LDRtg*0) / TotalGames
    else: 
        AdjORtg = (WORtg*NumWins + LORtg*NumLoss) / TotalGames
        AdjDRtg = (WDRtg*NumWins + LDRtg*NumLoss) / TotalGames #adjusted offense and defensive ratings based on total games 
    
    
 #strength of schdedule 
    #ratings 
    #out = pd.DataFrame({'Season': season,'Team1': TeamID, 'Wins': [NumWins], 'Losses': [NumLoss]
                       # ,'TotalGames': TotalGames,'TotalScored': TotalScored, 'TotalAllowed': TotalAllowed,
                       # 'eFG': eFG, 'TO%': TO_Percent, 'FT': FTrate, 
                       #'ORp': ORp, 'AdjORtg': AdjORtg, 'AdjDRtg': AdjDRtg})
   
    
    return  [NumWins, NumLoss, TotalGames, TotalScored, TotalAllowed,eFG, TO_Percent, FTrate, ORp, AdjORtg, AdjDRtg]

## team has to be in D1 
stats_per_season(1101, 2003): 
```
